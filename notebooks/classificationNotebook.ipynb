```python
# read image url list from datalack
df = spark.sql("SELECT * FROM InFreGen.task WHERE state = '0'")
picPaths = df.select("curPicPath").collect()
```

```python
# load immage as a kind of dataframe
image_df = spark.read.format("image").load("Files/images/23/1_small_cutted.jpg")
image_df.select("image.origin", "image.width", "image.height").show(truncate=False)
```

```python
import torch
import mlflow.pytorch
from PIL import Image
from torchvision import transforms

# Load the model from MLflow
model_uri = "models:/classification/1"  # Use the correct run_id
model = mlflow.pytorch.load_model(model_uri)

# Set model to evaluation mode
model.eval()

# Preprocess input image
preprocess = transforms.Compose([
    transforms.Resize(224),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
``
for picPath in picPaths:
    # Load an image
    image = Image.open(picPath)
    input_tensor = preprocess(image)
    input_batch = input_tensor.unsqueeze(0)  # Add batch dimension

    # Run inference
    with torch.no_grad():
        output = model(input_batch)

    # Get the predicted class
    _, predicted_class = torch.max(output, 1)
    print("Predicted class index:", predicted_class.item())
```
